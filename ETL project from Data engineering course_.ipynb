## Extract


```python
def extract_from_json(file_to_process):
    dataframe = pd.read_json(file_to_process)
    return dataframe
```

```python
columns=['Name','Market Cap (US$ Billion)']
```

```python
def extract():
    return extract_from_json('bank_market_cap_1.json')[0]
```

```python
# Get the exchange rate for later use
ex_rates = pd.read_csv("exchange_rates.csv", index_col = 0)
ex_rates

exchange_rate = ex_rates[ex_rates.index == 'GBP'].iloc[0]['Rates']
exchange_rate
```

## Transform


```python
def transform(data):
    
    # 1.
    data['Market Cap (US$ Billion)'] = data['Market Cap (US$ Billion)'] * exchange_rate
    
    # 2. 
    data = data.round({'Market Cap (US$ Billion)': 3})
        
    # 3.
    data = data.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'})
    
    return data
```

## Load

```python
def load(data):
    data.to_csv('bank_market_cap_gbp.csv', index=False)  
```

## Logging Function


```python
def log(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S' # Year-Monthname-Day-Hour-Minute-Second
    now = datetime.now() # get current timestamp
    timestamp = now.strftime(timestamp_format)
    with open("bank_market_cap_gbp.csv","a") as f:
        f.write(timestamp + ',' + message + '\n')
```
